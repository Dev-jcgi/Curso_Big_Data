# üìò Soluci√≥n del Estudio de Caso: An√°lisis de Comportamiento en Retail Digital

## 1. Resumen del Proyecto
**Tema:** An√°lisis del comportamiento de compra mediante Big Data.  
**Tecnolog√≠a:** Procesamiento Distribuido (Apache Spark) y Anal√≠tica Avanzada.  
**Contexto:** Optimizaci√≥n de una plataforma tipo Instacart con presencia internacional.

---

## 2. An√°lisis del Problema (üîç Resoluci√≥n)

Para abordar el volumen masivo de datos, la soluci√≥n t√©cnica debe estructurarse en una arquitectura de **Pipeline de Datos** escalable:

* **Ingesta:** Carga de archivos CSV distribuidos desde un Data Lake (S3/Azure Blob Storage).
* **Procesamiento:** Uso de **Apache Spark** para transformar datos mediante RDDs o DataFrames, permitiendo el procesamiento en paralelo en un cl√∫ster.
* **Almacenamiento:** Los resultados procesados se mueven a un Data Warehouse (como BigQuery o Snowflake) para el consumo de BI.



---

## 3. Respuestas a Preguntas Detonadoras (üß†)

### üß† T√©cnicas
* **An√°lisis Exploratorio Clave:** El an√°lisis de **Market Basket Analysis** (Reglas de Asociaci√≥n) y el an√°lisis de **Cohortes** para medir la retenci√≥n y frecuencia de recompra.
* **Preprocesamiento:** Es vital para manejar valores nulos (pedidos cancelados), normalizar nombres de productos y eliminar duplicados que sesgan las m√©tricas de ventas.
* **Spark vs. Pandas:** Pandas carga los datos en la memoria RAM de una sola m√°quina (l√≠mite f√≠sico). Spark distribuye los datos en un cl√∫ster, permitiendo procesar Terabytes que no cabr√≠an en una sola computadora.
* **Productos juntos:** Se utiliza el algoritmo **FP-Growth** o **Apriori** para encontrar conjuntos de elementos frecuentes y calcular la probabilidad de que, si compras el producto A, compres el B.

### üöÄ Innovaci√≥n y Tendencia
* **Recomendaciones en tiempo real:** Integrando **Spark Streaming** con Kafka, se pueden capturar los clics actuales del usuario y comparar con el modelo hist√≥rico para sugerir productos antes de que finalice el pago.
* **Personalizaci√≥n:** Mediante Big Data se crean perfiles "Buyer Persona" din√°micos, ajustando precios y promociones seg√∫n la elasticidad de precio detectada en cada usuario.

### üåç Internacionalizaci√≥n y √âtica
* **Desaf√≠os:** Diferencias en divisas, formatos de fecha, normativas fiscales y, sobre todo, leyes de protecci√≥n de datos locales (como GDPR en Europa vs. leyes menos restrictivas en otros pa√≠ses).
* **√âtica y Privacidad:** El riesgo principal es la **re-identificaci√≥n** (anonimato insuficiente). Se debe mitigar mediante t√©cnicas de **Privacidad Diferencial** y encriptaci√≥n de datos sensibles.

---

## 4. Propuesta de Soluci√≥n T√©cnica (Ejemplo de L√≥gica)

Para identificar los productos m√°s vendidos y mejorar la log√≠stica:

1.  **Agregaci√≥n por Hora/D√≠a:**
    $$Ventas(h) = \sum_{i=1}^{n} Pedidos_{i}$$

2.  **M√©trica de Recompra (Churn Rate):**
    Identificar qu√© porcentaje de usuarios vuelve a comprar en un periodo de 30 d√≠as para evaluar la lealtad.

## 5.1. Preprocesar los datos eliminando variables irrelevantes, imputando valores faltantes y limpiando los nombres de los productos.

```python
# Instalaci√≥n de m√≥dulos (solo necesario en entornos limpios como Colab)
# En entornos locales con entorno virtual ya configurado, se puede omitir.
!pip install pandas
!pip install matplotlib
!pip install networkx
```

```python
# Importar m√≥dulos
# Procesamiento de datos
import pandas as pd
# Visualizaci√≥n
import matplotlib.pyplot as plt
# Grafos para co‚Äëocurrencias de productos
import networkx as nx
from itertools import combinations
from collections import Counter
```

```python
# Carga de datos desde archivos CSV
# Cada DataFrame representa una entidad del modelo (ordenes, productos, etc.)
df_department = pd.read_csv('data/departments.csv')
df_orders = pd.read_csv('data/orders.csv')
df_products = pd.read_csv('data/products.csv')
df_order_products = pd.read_csv('data/order_products__prior.csv')
```

```python
# Eliminar atributos que no se usar√°n en este an√°lisis
# (reducimos ruido y tama√±o del dataset)
df_orders = df_orders.drop('eval_set', axis=1)
df_orders = df_orders.drop('days_since_prior_order', axis=1)
df_products = df_products.drop('aisle_id', axis=1)
df_order_products = df_order_products.drop('reordered', axis=1)
```

```python
# Unir los datos en una sola tabla anal√≠tica
# Orden ‚Üí productos ‚Üí cat√°logo ‚Üí departamentos
# Resultado: tabla con contexto completo de cada producto comprado

data = df_orders.merge(df_order_products, on="order_id") \
               .merge(df_products, on="product_id") \
               .merge(df_department, on="department_id")
```

```python
# Validaci√≥n de calidad: valores nulos
# Esto ayuda a decidir si imputar, eliminar o corregir registros
valor_imputar = data.isnull().sum()
print(valor_imputar)
```

No hay valores NULOS

```python
# Verificar nombres de productos duplicados o inconsistentes
# Un conteo alto puede indicar ruido o inconsistencias en el cat√°logo
valor_duplicados = data['product_name'].value_counts()
print(valor_duplicados)
```

```python
# Limpieza de nombres de productos
# - eliminar espacios extra
# - convertir a min√∫sculas para evitar duplicados por may√∫sculas

data['product_name'] = data['product_name'].str.strip()
data['product_name'] = data['product_name'].str.lower()
```

```python
# Revalidar duplicados tras limpieza
valor_duplicados = data['product_name'].value_counts()
print(valor_duplicados)
```

### 5.2.2 Frecuencia de compras por d√≠a

Convertimos el d√≠a num√©rico a nombre para visualizarlo.

### 5.2.3 Frecuencia de compras por hora

Convertimos la hora num√©rica a formato AM/PM para la visualizaci√≥n.

## 5.3 Productos m√°s vendidos (global)


```python
# Top productos m√°s vendidos (global)
# 1) Tomar nombre de producto
# 2) Contar ocurrencias
# 3) Extraer top N
nombre_prod = data[['product_id', 'product_name']]
cant_prod = nombre_prod['product_name'].value_counts()
top_prod = cant_prod.head(5)
```

```python
# Gr√°fico de los productos m√°s vendidos
# Se usa barra horizontal para facilitar lectura de nombres largos
top_prod.plot(kind='barh', edgecolor='black', color='skyblue')
plt.xlabel('N√∫mero de veces vendido')
plt.ylabel('Nombre del producto')
plt.title('Los 5 productos m√°s vendidos')
# Invertir eje Y: el m√°s vendido queda arriba
plt.gca().invert_yaxis()
plt.show()
```

## 5.4 Producto m√°s comprado por departamento


```python
# Producto m√°s comprado por departamento
# 1) Agrupar por departamento y producto
# 2) Tomar el top 1 por cada departamento

top_prod_x_dep = data.groupby('department')['product_name'] \
    .value_counts() \
    .groupby(level=0) \
    .nlargest(1)

# Restablecer el primer nivel de √≠ndice
# (para que el resultado sea un Series con el departamento como √≠ndice principal)
top_prod_x_dep = top_prod_x_dep.reset_index(level=0, drop=True)
```

```python
# Gr√°fico: producto m√°s comprado por departamento
plt.figure(figsize=(10,10))
top_prod_x_dep.plot(kind='barh', edgecolor='black', color='skyblue')
plt.title('Producto m√°s comprado por departamento')
plt.xlabel('Cantidad de pedidos')
plt.ylabel('Departamento / Producto')
plt.gca().invert_yaxis()
plt.show()
```

## 5.5 Productos comprados juntos (co‚Äëocurrencia)


```python
# Co‚Äëocurrencia de productos (Market Basket simple)
# 1) Lista de productos por pedido
# 2) Combinar pares dentro de cada pedido
# 3) Contar frecuencia de pares

orden_prod = data.groupby('order_id')['product_name'].apply(list)
comb_prod = Counter()

for products in orden_prod:
    comb_prod.update(combinations(products, 2))

# Top 5 combinaciones m√°s frecuentes
top_comb_prod = comb_prod.most_common(5)
```

```python
# Grafo de co‚Äëocurrencia (productos comprados juntos)
G = nx.Graph()

# A√±adir nodos y aristas con peso (frecuencia)
for (p1, p2), count in top_comb_prod:
    G.add_node(p1)
    G.add_node(p2)
    G.add_edge(p1, p2, weight=count)

# Dibujar el grafo
plt.figure(figsize=(10,10))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=2000, font_size=10)
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.title('Los 5 productos m√°s comprados juntos')
plt.show()
```
