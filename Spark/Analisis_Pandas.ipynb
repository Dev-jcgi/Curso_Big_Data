### üìä Pandas vs SQL

| Pandas | SQL Equivalente |
|--------|----------------|
| `df[df['col'] > 10]` | `SELECT * FROM df WHERE col > 10` |

| `pd.merge(df1, df2, on='key')` | `SELECT * FROM df1 JOIN df2 ON df1.key = df2.key` |
| `df.sort_values('col')` | `SELECT * FROM df ORDER BY col` |

El **preprocesamiento** es una etapa cr√≠tica en cualquier proyecto de an√°lisis de datos. Incluye:

**¬øPor qu√© es importante?**
- üéØ Mejora la calidad de los resultados
- ‚ö° Reduce el tiempo de procesamiento

## 5.5. Identificar los productos que m√°s se compran juntos, es decir, aquellos que aparecen en los mismos pedidos con mayor frecuencia, y visualizar los resultados utilizando un gr√°fico de red.

---

````python
"""
Identificaci√≥n de productos que se compran juntos (Market Basket Analysis)
An√°lisis de asociaci√≥n de productos para encontrar patrones de compra conjunta
Proceso:
1. Agrupar productos por pedido
2. Generar todas las combinaciones posibles de pares de productos por pedido
3. Contar la frecuencia de cada combinaci√≥n
4. Obtener las 5 combinaciones m√°s frecuentes
"""
# Agrupar productos por pedido, creando una lista de productos para cada order_id
orden_prod = data.groupby('order_id')['product_name'].apply(list)

# Contador para almacenar la frecuencia de combinaciones de productos
comb_prod = Counter()

# Iterar sobre cada pedido y generar combinaciones de pares de productos
for products in orden_prod:
    # combinations(products, 2) genera todas las combinaciones de 2 productos
    comb_prod.update(combinations(products, 2))

# Obtener las 5 combinaciones m√°s comunes
top_comb_prod = comb_prod.most_common(5)
````

````python
"""
Visualizaci√≥n: Grafo de red de productos comprados juntos
Crea una red donde:
- Nodos = Productos
- Bordes = Conexi√≥n entre productos comprados juntos
- Peso de los bordes = Frecuencia de compra conjunta

Este gr√°fico ayuda a visualizar las relaciones entre productos y 
puede ser √∫til para estrategias de cross-selling o ubicaci√≥n en tienda
"""
# Crear un grafo vac√≠o
G = nx.Graph()

# Agregar nodos (productos) y bordes (relaciones) al grafo
for (p1, p2), count in top_comb_prod:
    G.add_node(p1)  # Agregar primer producto como nodo
    G.add_node(p2)  # Agregar segundo producto como nodo
    G.add_edge(p1, p2, weight=count)  # Agregar conexi√≥n con peso (frecuencia)

# Crear la visualizaci√≥n del grafo
plt.figure(figsize=(10,10))

# Calcular posiciones de los nodos usando spring layout (distribuci√≥n de fuerzas)
pos = nx.spring_layout(G)

# Dibujar el grafo con nodos y etiquetas
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=2000, font_size=10)

# Obtener y mostrar las etiquetas de los bordes (pesos/frecuencias)
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)

plt.title('Los 5 productos m√°s comprados juntos')
plt.show()
````

### Interpretaci√≥n: Productos Comprados Juntos (Market Basket Analysis)

**Hallazgos clave:**
- Existen patrones claros de compra conjunta entre productos espec√≠ficos
- Las conexiones en el grafo muestran la fuerza de asociaci√≥n (n√∫meros en las aristas)
- Algunos productos act√∫an como "conectores" apareciendo en m√∫ltiples combinaciones
